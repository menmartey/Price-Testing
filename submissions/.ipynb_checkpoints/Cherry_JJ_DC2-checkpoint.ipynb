### Okay!  That means that we can take the results straight from the grouped dataframe

```python
reduced_df = test_df.groupby(by=['test','converted']).count().reset_index()
# sales fraction for the test case
sales_frac_test = reduced_df.loc[(reduced_df['test']==1) & (reduced_df['converted']==1),'user_id'].values[0]/\
                  (reduced_df.loc[(reduced_df['test']==1) & (reduced_df['converted']==1),'user_id'].values[0] + \
                   reduced_df.loc[(reduced_df['test']==1) & (reduced_df['converted']==0),'user_id'].values[0])
# sales fraction for the null case
sales_frac_null = reduced_df.loc[(reduced_df['test']==0) & (reduced_df['converted']==1),'user_id'].values[0]/\
                  (reduced_df.loc[(reduced_df['test']==0) & (reduced_df['converted']==1),'user_id'].values[0] + \
                   reduced_df.loc[(reduced_df['test']==0) & (reduced_df['converted']==0),'user_id'].values[0])
```

```python
# These will be the total income per customer for the null and test cases
print ("Income per customer for the test case = $",59.*sales_frac_test)
print ("Income per customer for the null case = $",39.*sales_frac_null)
```

# Now Let's look at how the data looks on a map of the US

### I'll start by making a combined dataframe using the test data and the user data.

```python
combined_df = pd.merge(test_df, user_df, left_on='user_id', right_on='user_id')
```

```python
combined_df.head()
```

```python
# set up a geo dataframe to do the plotting
import geopandas as gpd
from shapely.geometry import Point,Polygon
import contextily
combined_df['geometry'] = combined_df[['long','lat']].apply(Point, axis=1) #takes the lat and long data and turns it into a geometry column
geo_df = gpd.GeoDataFrame(combined_df) #turn into a geo-data frame
geo_df.crs = {'init':'epsg:4269'} #sets the coordinate reference system to lat long
geo_df = geo_df.to_crs(epsg=3857) #change crs to pseudo mercator projection with units of meters
```

```python
fig,ax = plt.subplots(1,1,figsize=(25,25))
basemap, basemap_extent = contextily.bounds2img(*geo_df.total_bounds, zoom=5) #use contextily to grab a base map for the region spanned by the hoods polygons
ax.set_title(["zoomed in"], fontsize=20)
ax.imshow(basemap, extent=basemap_extent)
geo_df.plot('converted',ax=ax, cmap='viridis', marker='.', markersize=40)
```
